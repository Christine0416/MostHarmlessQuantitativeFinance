# 主要是用来演示和转我们以前做的Matlab的历史波动率的代码


```python
#-*- coding:utf-8 -*-
import numpy as np
import pandas as pd
import scipy.stats as ss
import matplotlib.pyplot as plt
from scipy.optimize import fmin_slsqp
%matplotlib inline
df = pd.read_excel("SH600000.xlsx",parse_dates=[0])
df[:5]
df.index=df.pop('Date') 

df.plot(grid=True,figsize=(10,6))
df.head()
```

```python
def CtCHV(data, N, scal, out = None):
###---------------
# input: data : format must be PANDAS
#        N : historial window
#        out: assume using log diff returns
# output std_CtC
###-----------
    if out is None:
        dfrets =scal*np.log(data/data.shift(1)).dropna()
        print('running log diff returns')
    else:
        dfrets =scale*data.pct_change().dropna()
        print('running pct_change returns')
   
    T = np.size(dfrets,0)
    print(T)
    std_CtC=np.zeros((T-N,1))
    for t in range(0,T-N):
        std_CtC[t]=dfrets[t:t+N].std()
    
    return std_CtC
```

```python
def GermanKlassHV(HDD, LDD, ODD, CDD, N, scal):
    un = scal*np.log(HDD/ODD)
    dn = scal*np.log(LDD/ODD)
    cn = scal*np.log(CDD/ODD)
    C1=0.511
    C2=0.019
    C3=0.385
    
    T = np.size(HDD,0)
    std_gk=np.zeros((T-N,1))
    for i in range(0,T-N):
        prt1 = (C1/N)*(np.sum((un[i:i+N]-dn[i:i+N])**2))
        pprt1 = cn[i:i+N]*(un[i:i+N]+dn[i:i+N])
        pprt2 = 2*un[i:i+N]*dn[i:i+N]
        prt2 = -(C2/N)*np.sum(pprt1+pprt2)
        prt3 = -(C3/N)*np.sum(cn[i:i+N]**2)
        std_gk[i]=np.sqrt(prt1+prt2+prt3)
        
    return std_gk
```

```python
def parkinsonHV(HDD, LDD, ODD, N, scal):
    un = scal*np.log(HDD/ODD)
    dn = scal*np.log(LDD/ODD)

    T = np.size(HDD,0)
    std_park=np.zeros((T-N,1))
    for i in range(0,T-N):
        std_park[i]=np.sqrt((1/(4*N*np.log(2)))*np.sum((un[i:i+N]-dn[i:i+N])**2))
        
    return std_park
```

```python
def RogersSatchellHV(HDD, LDD, ODD, CDD, N, scal):
    un = scal*np.log(HDD/ODD)
    dn = scal*np.log(LDD/ODD)
    cn = scal*np.log(CDD/ODD)
    
    T = np.size(HDD,0)
    std_rs=np.zeros((T-N,1))
    for i in range(0,T-N):
        tmp=un[i:i+N]*(un[i:i+N]-cn[i:i+N])+dn[i:i+N]*(dn[i:i+N]-cn[i:i+N]);
        std_rs[i]=np.sqrt(np.mean(tmp));
        
    return std_rs
```

```python
def gjr_garch_likelihood(parameters, data, sigma2, out = None):
    mu = parameters[0]
    omega = parameters[1]
    alpha = parameters[2]
    gamma = parameters[3]
    beta = parameters[4]
    
    T = np.size(data,0)
    #print('What is T=',T)
    eps = data - mu
    # Data and Sigma2 are T by 1 vectors
    for t in range(1,T):
        sigma2[t]=(omega+alpha*eps[t-1]**2+gamma*eps[t-1]**2 * (eps[t-1]<0)+beta*sigma2[t-1])
        
    logliks = 0.5*(np.log(2*np.pi)+np.log(sigma2)+eps**2/sigma2)
    loglik = np.sum(logliks)
    
    if out is None:
        return loglik
    else:
        return loglik, logliks, np.copy(sigma2)

def gjr_constraint(parameters, data, sigma2, out=None):
    alpha = parameters[2]
    gamma = parameters[3]
    beta = parameters[4]
    return np.array([1-alpha-gamma/2-beta])
```

```python
def garch_likelihood(parameters, data, sigma2, out = None):
    mu = parameters[0]
    alpha = parameters[1]
    beta = parameters[2]
    
    T = np.size(data,0)
    eps = data - mu
    # Data and Sigma2 are assumed as T by 1 vectors
    for t in range(1,T):
        sigma2[t]=(alpha*eps[t-1]**2+beta*sigma2[t-1])
        
    logliks = 0.5*(np.log(2*np.pi)+np.log(sigma2)+eps**2/sigma2)
    loglik = np.sum(logliks)
    
    if out is None:
        return loglik
    else:
        return loglik, logliks, np.copy(sigma2)

def garch_constraint(parameters, data, sigma2, out=None):
    alpha = parameters[1]
    beta = parameters[2]
    return np.array([1-alpha-beta])
```

# 下面是单独运行 Garch 1 1和所有其他的 HV 波动率

```python
N=252


HDD=np.array(df.HDD)
LDD=np.array(df.LDD)
ODD=np.array(df.ODD)
CDD=np.array(df.CDD)


std_CtC = CtCHV(df.CDD,N)
std_park = parkinsonHV(HDD,LDD,ODD, N)
std_gk_v2 = GermanKlassHV(HDD,LDD,ODD,CDD, N)
std_rs = RogersSatchellHV(HDD,LDD,ODD,CDD, N)

# or  using
#std_CtC = CtCHV(df.CDD,N)
#std_gk_v2 = GermanKlassHV(df.HDD,df.LDD,df.ODD,df.CDD, N)


df.CDD

dfrets = (df.CDD).pct_change().dropna() 
#dfrets =np.log(df.CDD/df.CDD.shift(1)).dropna() #log diff returns #comfirm many times
dfrets

mean_rts = dfrets.mean()
var_rts =dfrets.var();


#demo volatilities with given parameters
startingVals = np.array([mean_rts,.09,.90]) ##change

T=len(dfrets);

finfo=np.finfo(np.float64)

bounds  =[(-10*mean_rts,10*mean_rts),(0.0,1.0),(0.0,1.0)]

sigma2 = np.ones(T)*(var_rts)

args = (np.asarray(dfrets),sigma2)

estimates = fmin_slsqp(garch_likelihood, startingVals, f_ieqcons=garch_constraint, bounds=bounds, args =args)

print('Initial Values=',startingVals)
print('Estimated Values=',estimates)

analized=1 # or we should set to 252
sigma2 = np.ones(T)*(var_rts)
loglik, logliks, sigma2final = garch_likelihood(estimates,np.array(dfrets), sigma2,out=True)

garch_vol=np.sqrt(analized*sigma2final)

ested_vol = pd.DataFrame(garch_vol,index=dfrets.index,columns=['estimated vols'])

plt.figure(figsize=(12,7),dpi=980)
plt.grid(True)

plt.plot(std_CtC,label='CtC HV')
plt.plot(std_rs,label='RogersSatchell HV')
plt.plot(std_park,label='parkinson HV')
plt.plot(std_gk_v2,label=' GermanKlass HV')


plt.plot(garch_vol[252:],label=' Gjr Garch Vol')
plt.title('Histirical Volatilites')
plt.legend(loc = 'upper left') #must show up otherwise label 'xxx' would show up
plt.xlabel('Date',fontsize=20)
plt.ylabel('Volatility')

ested_vol.plot(grid='on',color = '#32CD32',title='SH60000 volatility with estimated garch')


print('ested loglik=',loglik)
ested_vol.describe()

#type(std_CtC)
#np.sum((std_gk_v2 - std_gk_v1)**2)
#std_gk_v2

#np.size(std_rs)

#data= df.CDD
#dfrets = 100* data.pct_change().dropna()
#T = np.size(data,0)
#std_CtC=np.zeros((T-N,1))
#for t in range(0,T-N):
#    std_CtC[t]=dfrets[t:t+N].var()
        
```
